In this notebook we will be diving into an anonymous organization's social media ad campagin. The data was accessed from a public Kaggle resource which can be accessed <a href='https://www.kaggle.com/datasets/loveall/clicks-conversion-tracking?resource=download'>here</a>. 

The file `KAG_conversion_data.csv` contains 1143 observations in 11 variables. Below is a description of each column in the dataset:

`ad_id`: an unique ID for each ad.

`xyz_campaign_id`: an ID associated with each ad campaign of XYZ company.

`fb_campaign_id`: an ID associated with how Facebook tracks each campaign.

`age`: age of the person to whom the ad is shown.

`gender`: gender of the person to whim the add is shown

`interest`: a code specifying the category to which the person’s interest belongs (interests are as mentioned in the person’s Facebook public profile).

`Impressions`: the number of times the ad was shown.

`Clicks`: number of clicks on for that ad.

`Spent`: Amount paid by company xyz to Facebook, to show that ad.

`Total conversion`: Total number of people who enquired about the product after seeing the ad.

`Approved conversion`: Total number of people who bought the product after seeing the ad.

<h1> Data Processing and Cleaning</h1><a id='DPC'></a>

In this section we will be taking an overview look at the data and imputing any null values.

```python
# Importing necessary libraries
import pandas as pd
import numpy as np

df = pd.read_csv('KAG_conversion_data.csv')

df.head()
```

```python
df.info()
df.describe()
```

It seems that there are no null values in the dataset and there are no obvious outliers in the features by looking at the quantile ranges. However, upon further inspection, some of the data is erroneous since there were no `Clicks` yet a non-zero `Total Conversion`. There cannot be a conversion if a user did not click on the ad.

```python
# change datatype of campaign id to object since it is not continous variable and is an identifying number/name
df['xyz_campaign_id'] = df['xyz_campaign_id'].astype(object, copy = False)

print(f"Number of Conversion with no Clicks: {len(df.loc[((df['Clicks'] == 0) & (df['Total_Conversion'] !=0)), ])}")
```

We can impute the click values based on their nearest neighbours using a KNN imputer. 

```python
from sklearn.impute import KNNImputer

# Make all 0 clicks NaN values to impute
df['Clicks'].replace(0, np.nan, inplace=True)

imputer = KNNImputer(n_neighbors=2)
df_imputed = pd.DataFrame(imputer.fit_transform(df.drop(columns=["gender", "age"])))

df["Clicks"] = df_imputed[5]
```

We have now imputed the `Clicks` column with the closest data points where the other columns are similar. 

The columns `ad_id`, `xyz_campaign_id` and `fb_campaign_id` are all identifying tags for the marketing campaigns. This information is redundant and therefore we will only keep `xyz_campaign_id`. Additionally, the `interest` column provides very little information to a users behaviour and whether they will perform a conversion or not, therefore we will drop this column as well.

```python
df = df.drop(columns=['ad_id', 'fb_campaign_id', 'interest'])

df.head()
```

<h1>Campaign Performance Comparison </h1> <a id='PERFORM'></a>

To correctly evaluate the performance of a campaign, we must look to some important Key Performance Indicators (KPIs) that allow for a more accurate comparison between campaigns even with large differences in spending. I have chosen to focus on three KPIs which are described below:

1. Conversion Rate (CR): ratio that measures the success of converting users from a larger group (such as website visitors or ad viewers) into taking a specific desired action (like making a purchase or signing up).

2. Cost Per Click (CPC): refers to the amount of money spent on average for each individual click on an advertisement or link.

3. Click-through rate (CTR): the ratio of the number of times users clicked on a specific link, advertisement, or call-to-action to the total number of times it was shown or presented, often used as a measure of engagement.

We will first create these KPIs from the existing data and visualize them.

```python
# feature engineering based on properties of 
df["CR"] = df["Total_Conversion"] / df["Impressions"]
df["CTR"] = df["Clicks"] / df["Impressions"]
df["CPC"] = df["Spent"] / df["Clicks"]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5))

axes = axes.flatten()

# Iterate through the columns of the DataFrame, excluding the first column
for i, col in enumerate(df.columns[-3:]):
    ax = axes[i]  # Get the current axis
    df[["xyz_campaign_id", col]].pivot_table(columns="xyz_campaign_id", values=col, aggfunc=np.mean).plot(kind="bar", ax=ax, legend=None)
    ax.set_title(f"Average {col} Between Campaigns".title())
    ax.tick_params(labelrotation=0)


handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=3)
axes[0].set_ylabel('Ratio')
plt.show()
```

Some observations when looking at the visualized averages in KPIs are:
- Campaign 936 was the most successful with a higher average CR, CTR, and lower CPC. 
- Campaign 916 follows a similar trend with relatively higher average CR and CTR with a lower CPC, although they were not as successful as campaign 936 there may not be a significant difference between the two groups.
- In contrast, although campaign 1178 spent far more than the other two campaigns as previously discussed, their average CR and CTR are much lower than campaigns 916 and 936.

Let's get into comparing the groups with statistical tests

<h1>A/B testing</h1><a id='stats'></a>

The primary objective of A/B testing for three distinct groups involves making pairwise comparisons between each combination of groups (A-B, A-C, B-C). This will allow us to determine if a statistically significant difference exist between these pairs or if they occur by chance. 

Since multiple comparisons are conducted together, there is an increased likelihood of erroneous positive conclusions (Type 1 error). This is known as the <i>multiple comparison problem</i>, which can lead to a misinterpretation of results due to the simultaneous nature of the inferences.

The Bonferroni correction technique can be employed to create stricter significance thresholds for rejections therefore accounting for the multiple inferences. The steps are as follows:

1. Assess the normality of residuals within each group using the Shapiro-Wilk test.
    a. If residuals are normally distributed, assess the homogeneity of variances among the groups using Levene's test.
        i. If variance homogeneity is determined, proceed with pair-wise T-tests.
        ii. If variance homogeneity is not determined, proceed with pair-wise Welch's tests.
    b. If residuals are not normally distributed, proceed with pair-wise Mann Whitney U tests (non-parametric test).
2. Apply the Bonferroni correction to draw conclusions from the tests.

Although the Bonferroni correction is frequently employed and straightforward, it is worth noting that this conservative approach can lead to a reduction in statistical power. Therefore, alternative methods such as the Tukey HSD test, multiple range test (MRT), and others might be considered. Nevertheless, given the considerable number of comparisons being performed, the use of the Bonferroni correction is advisable for the sake of simplicity and control.

For the significance level &alpha; = 0.05 but since we are applying the Bonferroni correction method, we divide the significance level by the number of pairwise comparisons, in our case 3. Therefore we will have an adjusted <b> &alpha; of 0.01667. </b>

<b>Residual Normality Check</b>

We will conduct the Shapiro-Wilk test with an &alpha; of 0.01667 and the following hypothesis statements:

$$H_0: \text{The distribution of the sample is close to normal.}$$
$$H_1: \text{The distribution of the sample is not close to normal.}$$

```python
from scipy.stats import shapiro

# Set the significance level (alpha) for the test
alpha = 0.01667  # Adjust this value according to your desired significance level

# Loop through each KPI column
for kpi in df.columns[-3:]:
    print(f"KPI: {kpi}")
    
    # Loop through unique values in the 'xyz_campaign_id' column
    for g in df.xyz_campaign_id.unique():
        # Perform the Shapiro-Wilk test on the subset of data for the current group and KPI
        _, p = shapiro(df.loc[df['xyz_campaign_id'] == g, kpi])
        
        # Compare the p-value to the significance level to make a decision
        if p < alpha:
            h = 'Reject Null Hypothesis'
        else:
            h = 'Fail To Reject the Null Hypothesis'
        
        # Print the results for the current group and KPI
        print(f"Group {g} p-value: {p}")
        print(f"Hypothesis: {h}\n")
```

Since all three KPIs have a p-value much lower than the significance threshold (&alpha; 0.01667), there is strong evidence that the groups' residuals are not normal. This does not statisfy the assumptions of a parametric test such as pair-wise T-test. Therefore, we must continue to a non-parametric test, in this case the Mann Whitney U test.

<b>Mann Whitney U test</b>

We will conduct the Mann Whitney U test with an &alpha; of 0.01667 and the following hypothesis statements:

$$H_0: \text{There is no difference (in central tendency) between the two groups in the population.}$$
$$H_1: \text{There is a difference (in central tendency) between the two groups in the population.}$$

```python
from scipy.stats import mannwhitneyu

# Define a function to perform Mann-Whitney U test and print results
def compare(kpi, alpha):
    print(f"KPI: {kpi}")
    
    # Perform Mann-Whitney U tests for three pairs of groups
    _, p1 = mannwhitneyu(df.loc[df['xyz_campaign_id'] == 916, kpi], df.loc[df['xyz_campaign_id'] == 936, kpi])
    _, p2 = mannwhitneyu(df.loc[df['xyz_campaign_id'] == 936, kpi], df.loc[df['xyz_campaign_id'] == 1178, kpi])
    _, p3 = mannwhitneyu(df.loc[df['xyz_campaign_id'] == 1178, kpi], df.loc[df['xyz_campaign_id'] == 916, kpi])
    
    # Compare p-values to the significance level and determine hypothesis conclusion
    h1 = "Reject Null Hypothesis" if p1 < alpha else "Fail to Reject Null Hypothesis"
    h2 = "Reject Null Hypothesis" if p2 < alpha else "Fail to Reject Null Hypothesis"
    h3 = "Reject Null Hypothesis" if p3 < alpha else "Fail to Reject Null Hypothesis"
    
    # Print the results for each pair of groups and their hypotheses
    print(f"Group 916-936 p-value: {p1}\nHypothesis: {h1}\n\nGroup 936-1178 p-value: {p2}\nHypothesis: {h2}\n\nGroup 1178-916 p-value: {p3}\nHypothesis: {h3}\n")

# Set the significance level (alpha) for the tests
alpha = 0.05  # Adjust this value according to your desired significance level

# Loop through each KPI column and perform the test
for kpi in df.columns[-3:]:
    compare(kpi, alpha)
```

From the Mann Whitney U test, we see that there is no statistically significant difference between campaigns 916 and 936 for all three KPIs and any difference between them is due to chance. In contrast, both campaigns 916 and 936 have a statistically significant difference compared to campaign 1178 for all three KPIs.

# Conclusion <a id='fin'></a>
With regards to the company's ad campagin this means that any changes in spending for campagins 916 or 936 would not result in signficiant impacts on conversion rate, engagement (CTR) and cost per click over the other. Whereas, the significant difference between 916 and 936 to 1178 shows that, relative to their spending, the two campaigns have performed signficiantly better in terms of conversion, engagement and cost-efficiency than 1178 despite the higher investment in 1178.

We can determine some actionable solutions from these results:
- Since campagins 916 and 936 have had better performance in CR and CTR, the company should consider up-scaling their investment. 
- Given the large budget, 1178 was able to generate many conversions and impressions, however, the efficiency is far lower than the other two campaigns with a lower budget. The company should further investigate the factors contributing to the lack of engagement and conversions, especially since it was exposed to a much larger audience. 
- Since the most interested audiences are from the age group of 30-34, the company should further target this age group for potentially higher user engagement and interactions. 
- The other age groups should be investigated to understand the lack of engagement and possibly develop other ad campaigns to target them.
